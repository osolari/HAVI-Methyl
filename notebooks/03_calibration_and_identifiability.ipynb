# 3. Conformal calibration and identifiability

This notebook explores the conformal wrappers (Sec. 8) and the identifiability stress test (Sec. 11.4).

  - Split-conformal: Prop. 4 marginal coverage.
  - Mondrian: per-stratum coverage (Sec. 8.4).
  - Identifiability: prior leakage with no-VIB / VIB / VIB+mQTL (Sec. 7, 11.4).

```python
import sys, os
sys.path.insert(0, os.path.abspath('../src'))
import numpy as np
import havi_methyl as hm
from havi_methyl.calibration import gaussian_conformal_intervals
```

```python
rng = np.random.default_rng(20260429)
n_cal = 1000; n_test = 2000
means_cal = rng.normal(0, 1, n_cal); stds_cal = rng.uniform(0.5, 1.5, n_cal)
y_cal = means_cal + stds_cal * rng.standard_normal(n_cal)
means_test = rng.normal(0, 1, n_test); stds_test = rng.uniform(0.5, 1.5, n_test)
y_test = means_test + stds_test * rng.standard_normal(n_test)
lo, hi = gaussian_conformal_intervals(means_cal, stds_cal, y_cal, means_test, stds_test, alpha=0.10)
print('Empirical coverage:', round(((y_test >= lo) & (y_test <= hi)).mean(), 3))
print('Mean width:        ', round((hi - lo).mean(), 3))
```

```python
leak = hm.run_identifiability_stress_test(S=12, L=200, coverage=5.0, rng=rng)
print(leak.as_dict())
```

Notice that the simplified-Gaussian variant *under-covers* at low coverages (Sec. 11.3); production deployment uses the conformal wrapper above to recover the marginal guarantee of Prop. 4.
